```python
# Import Splinter and BeautifulSoup
import requests
from splinter import Browser
from bs4 import BeautifulSoup as soup
```

```python
# Set up Splinter
browser = Browser('chrome')
```

```python
# Visit the Mars news site
url = 'https://www.spotrac.com/nfl/chicago-bears/cap/'
browser.visit(url)

# Optional delay for loading the page
browser.is_element_present_by_css('div.list_text', wait_time=5)
```

```python
# Scrape the website
html = browser.html

# Create a Beautiful Soup object
chicago_soup = soup(html, 'html.parser')

# Select one news article with select_one()
chicago_soup.text

active_player_elem = chicago_soup.find_all('td', class_='player')
active_player_elem[0]
```
